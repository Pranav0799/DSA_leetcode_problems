```python
import requests
import re
import io
from io import StringIO
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
url = ' https://urlhaus.abuse.ch/downloads/csv_recent/'
```

```python
response = requests.get(url)
csv_text = response.text
```

```python
preprocessed_text = csv_text.lower()
```

```python
preprocessed_text = re.sub(r'[^\w\s]', '', preprocessed_text)
```

```python
csv_data = pd.read_csv(StringIO(preprocessed_text))
# loaded preprocessed csv data into pandas dataframe
tokens = word_tokenize(preprocessed_text)
```

```python
import json
with open('/new 3.json','r') as file:
  json_data = json.load(file)
  #/new 3.json
#ids = [entry['ID'] for entry in json_data]
#titles = [entry['title'] for entry in json_data]
titles = []
for entry in json_data:
  if 'title' in entry:
    titles.append(entry['title'])
```

```python
all_text = ' '.join(titles)
tokens2 = word_tokenize(all_text)
```
